```python
import time
import matplotlib.pyplot as plt
import random
import sys
sys.setrecursionlimit(20000)

# Kode sebelumnya dengan penyesuaian untuk pendekatan rekursif
class Sesi:
    def __init__(self, mata_kuliah, hari, kapasitas):
        self.mata_kuliah = mata_kuliah
        self.hari = hari
        self.kapasitas = kapasitas
        self.mahasiswa = []
        self.jumlah_mahasiswa = 0

sesi_praktikum = [
    Sesi("StrukturData", "Senin", 20),
    Sesi("StrukturData", "Rabu", 20),
    Sesi("StrukturData", "Jumat", 20),
    Sesi("Algoritma", "Selasa", 20),
    Sesi("Algoritma", "Kamis", 20),
    Sesi("BasisData", "Senin", 20),
    Sesi("BasisData", "Rabu", 20),
    Sesi("BasisData", "Sabtu", 20)
]

def reset_sesi():
    for sesi in sesi_praktikum:
        sesi.mahasiswa = []
        sesi.jumlah_mahasiswa = 0

def get_kode_mata_kuliah(mata_kuliah):
    kode_map = {
        "StrukturData": 1,
        "Algoritma": 2,
        "BasisData": 3
    }
    return kode_map.get(mata_kuliah, 0)

def jadwalkan_mahasiswa_iteratif(daftar_mahasiswa):
    for id_mahasiswa, mata_kuliah_list in daftar_mahasiswa:
        for mata_kuliah in mata_kuliah_list:
            kode_mk = get_kode_mata_kuliah(mata_kuliah)
            for sesi in sesi_praktikum:
                if (get_kode_mata_kuliah(sesi.mata_kuliah) == kode_mk and
                        sesi.jumlah_mahasiswa < sesi.kapasitas):
                    sesi.mahasiswa.append(id_mahasiswa)
                    sesi.jumlah_mahasiswa += 1
                    break

def jadwalkan_mahasiswa_rekursif(daftar_mahasiswa, index=0):
    if index >= len(daftar_mahasiswa):
        return True

    id_mahasiswa, mata_kuliah_list = daftar_mahasiswa[index]
    for mata_kuliah in mata_kuliah_list:
        kode_mk = get_kode_mata_kuliah(mata_kuliah)
        for sesi in sesi_praktikum:
            if (get_kode_mata_kuliah(sesi.mata_kuliah) == kode_mk and
                    sesi.jumlah_mahasiswa < sesi.kapasitas):
                sesi.mahasiswa.append(id_mahasiswa)
                sesi.jumlah_mahasiswa += 1
                break
    return jadwalkan_mahasiswa_rekursif(daftar_mahasiswa, index + 1)

# Fungsi untuk mengukur waktu eksekusi
def measure_time(func, *args):
    start_time = time.time()
    func(*args)
    end_time = time.time()
    return end_time - start_time

# Membuat data mahasiswa untuk pengujian
def generate_random_data(num_students):
    mata_kuliah_list = ["StrukturData", "Algoritma", "BasisData", "Sistem Operasi", "Basis Data", "Jaringan Komputer", "Kecerdasan Buatan", "Pemrograman Web", "Keamanan Siber", "Robotika", "Pemrosesan Paralel"]
    data = []
    for i in range(1, num_students + 1):
        mata_kuliah = random.sample(mata_kuliah_list, random.randint(1, 3))
        data.append((i, mata_kuliah))
    return data

# Eksperimen dan plotting
def run_experiment():
    student_counts = [10, 50, 100, 500, 1000, 5000, 6000, 7000, 8000, 9000, 10000]
    iterative_times = []
    recursive_times = []

    for count in student_counts:
        data = generate_random_data(count)

        reset_sesi()
        iterative_times.append(measure_time(jadwalkan_mahasiswa_iteratif, data))

        reset_sesi()
        recursive_times.append(measure_time(jadwalkan_mahasiswa_rekursif, data))

    plt.plot(student_counts, iterative_times, label="Iteratif", marker="o")
    plt.plot(student_counts, recursive_times, label="Rekursif", marker="o")
    plt.title("Perbandingan Waktu Eksekusi: Iteratif vs Rekursif")
    plt.xlabel("Jumlah Mahasiswa")
    plt.ylabel("Waktu Eksekusi (detik)")
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == "__main__":
    run_experiment()
```
